```python
import pandas as pd
import requests
```

```python
ca=pd.read_excel('./천안시 사망자 데이터.xlsx')
ca
```

```python
ca = ca[ca['읍면동'].str.contains('천안시')]
```

```python
# 경도 위도 추출
lenght=len(ca)
loca=[]
for i in range(0,lenght):
    x=ca.iloc[i]['경도']
    y=ca.iloc[i]['위도']
    loca.append([x,y])
    
```

```python
API_KEY = '5744b58eb5dac17a561b8e1983fe4440'

category=['CS2', 'PK6', 'MT1', 'OL7','BK9', 'CT1', 'PO3', 'FD6', 'HP8', 'PM9']
name=['편의점','주차장','대형마트','주유소,충전소','은행','문화시설','공공기관','음식점','병원','약국']


length=len(category)
for j in range(lenght):
	cate=category[j]
	count=[]
	for i in loca:
		x,y=i
		response = requests.get(f"https://dapi.kakao.com/v2/local/search/category.json?category_group_code={cate}&x={x}&y={y}&radius=100" , headers={'Authorization': f'KakaoAK {API_KEY}'})
		data=response.json()
		count.append(data['meta']['total_count'])
	title=name[j]
	ca[title]=count
ca
```
